```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('toofailid/DATA_TEKSTIKOGU_puhastatud.xlsx', engine='openpyxl')
print(data.shape)
```

```python
label = data['Lahendi liik'].to_list()
tekst = data['text_puhas'].to_list()
```

```python
tekstid = []
for i in range(len(label)):
    tekstid.append('__label__'+str(label[i]+ ' ' +tekst[i]))
```

```python
# Andmete tükeldamine treenimisandmestikuks ja testandmestikuks
from sklearn.model_selection import train_test_split
tekstid_train, tekstid_valid = train_test_split(tekstid, test_size=0.2, random_state=42)
```

```python
len(tekstid_train)
```

```python
len(tekstid_valid)
```

```python
textfile = open("toofailid/data.train.txt", "w", encoding="utf-8")
for element in tekstid_train:
    textfile.write(element + "\n")
textfile.close()
```

```python
textfile = open("toofailid/data.valid.txt", "w", encoding="utf-8")
for element in tekstid_valid:
    textfile.write(element + "\n")
textfile.close()
```

```python
import fasttext
model = fasttext.train_supervised(input="toofailid/data.train.txt")
```

```python
model.test("toofailid/data.valid.txt")
```
